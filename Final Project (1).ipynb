##**Project Title**

Exploratory Data Analysis of Used Cars Market in the UAE

##**Introduction**

The used car market in the UAE is one of the most active automotive markets due to high vehicle turnover, expatriate population, and demand for affordable mobility. Understanding pricing patterns, vehicle specifications, and consumer preferences is essential for buyers, sellers, and dealerships.
This project focuses on analyzing a large dataset of 10,001 used cars listed in the UAE, covering various brands, models, prices, mileage, engine specifications, and other attributes. By applying data analytics techniques, we aim to uncover meaningful insights and trends that influence used car prices in the UAE market

##**Dataset Description**

The dataset contains 10,001 rows and 12 columns, representing individual used car listings across the UAE.
Columns Used
-*Brand – Manufacturer of the car (Toyota, Nissan, BMW, etc.)*
-*Model – Specific model of the car (Camry, Patrol, X5, etc.)*
-*Description – Full car name (Brand + Model + Variant)*
-*Price (AED) – Selling price of the used car*
-*Year – Manufacturing year*
-*Mileage (KM) – Distance driven by the car*
-*Cylinder – Number of engine cylinders*
-*Fuel_Type – Petrol / Diesel / Hybrid*
-*Transmission – Automatic / Manual*
-*Colour – Exterior color*
-*Body_Type – Sedan, SUV, Hatchback, etc.*
-*Location – Emirate or city where the car is listed*

##**Objective of the Project**

The main objective of this project is to analyze the UAE used car market and identify factors that influence car prices.
Specific objectives include:
To understand how price varies with mileage, year, and brand
To identify most popular car brands and models in the UAE
To analyze the impact of engine cylinders and transmission type on price
To study price differences across body types and fuel types
To provide data-driven insights useful for buyers and sellers

***PRICE BY MANUFACTURING YEAR***

```python
# This code groups the dataset by manufacturing year and calculates the total sales price for each year.
# -`groupby("Year")` organizes the data based on the vehicle manufacturing year.
# - `["Price"].sum()` computes the total price of vehicles for each year.
# - `reset_index()` converts the grouped data into a structured DataFrame.
# - The resulting table shows year-wise total sales price.
year_avg_price=data.groupby("Year")["Price"].sum().reset_index(name="total sales price").sort_values(by="total sales price",ascending=False)
year_avg_price
```

```python
plt.figure(figsize=(12,6))
plt.title("yearly total sales")
plt.xlabel("year")
plt.ylabel("total sales amount")
plt.bar(year_avg_price["Year"],year_avg_price["total sales price"],color="seagreen",edgecolor="black")
plt.show()
```

**LOCATION WISE TRANSMISSION ANALYSIS**

```python
## Location-Wise Transmission Analysis

# This analysis examines how vehicle transmission types are distributed across different locations.

## Objective
# To understand the relationship between vehicle transmission type (e.g., Automatic, Manual) and location, and identify location-based preferences.

## Methodology
# - The dataset is grouped by **Location** and **Transmission**.
# - The number of listings for each transmission type in every location is calculated.
# - The result provides a comparative view of transmission usage across locations.

location_transmission_count=data.groupby(["Location","Transmission"]).size().reset_index(name="Total_Listing")
location_transmission_count                                                                                         
                                                                                         
```

```python
top_transmission_by_location = (
    location_transmission_count
    .sort_values(['Location', 'Total_Listing'], ascending=[True, False])
    .groupby('Location')
    .head(2)
)

top_transmission_by_location
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_locations = data['Location'].value_counts().head(6).index
filtered_data = data[data['Location'].isin(top_locations)]

plt.figure(figsize=(12,6))
sns.countplot(
    data=filtered_data,
    x='Location',
    hue='Transmission',
    order=top_locations
)

plt.xticks(rotation=45, fontsize=9)
plt.xlabel("Location")
plt.ylabel("Number of Listings")
plt.title("Transmission Type Distribution by Location")
plt.legend(title="Transmission", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Most Sold Car Colour Analysis**

```python
colour_count=data["Color"].value_counts().reset_index()
colour_count
```

**Colour-wise Sales Distribution by Location**

This visualization analyzes the distribution of the top 10 most sold car coloursacross different locations in the UAE. By combining colour and location data,the analysis highlights regional preferences for car colours in the used carmarket.

The dataset was grouped by car colour and location to calculate the total numberof listings for each combination. The top 10 most sold colours were selectedbased on overall listing count, and a bar chart was created to compare colourpreferences across locations.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10,5))
sns.barplot(
data=colour_count.head(10),
x='Color',
y='count',
hue=data['Location'],
palette='Set2')

plt.xticks(rotation=45, fontsize=9)
plt.xlabel("Car Colour", fontsize=11)
plt.ylabel("Number of Listings", fontsize=11)
plt.title("Top 10 Most Sold Car Colours", fontsize=13)
plt.tight_layout()
plt.show()

```
